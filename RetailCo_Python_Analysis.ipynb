This notebook demonstrates Python-based exploratory data analysis for the RetailCo project.
It includes data loading, cleaning, KPIs, and visualizations.


```

import pandas as pd
import matplotlib.pyplot as plt

# Example placeholder dataset
sales = pd.DataFrame({
    'Product': ['A','B','C','A','B'],
    'QuantitySold': [10, 5, 7, 12, 3],
    'Price': [20, 15, 10, 20, 15],
    'Date': pd.date_range(start='2024-01-01', periods=5)
})

sales.head()
```

```

# Calculate revenue
sales['Revenue'] = sales['QuantitySold'] * sales['Price']
total_revenue = sales['Revenue'].sum()
total_revenue
```

```

# Monthly revenue trend
monthly = sales.groupby(sales['Date'].dt.to_period('M'))['Revenue'].sum()

plt.figure(figsize=(6,4))
monthly.plot(marker='o')
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.tight_layout()
plt.show()
```
